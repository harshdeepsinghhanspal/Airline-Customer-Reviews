```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
review = pd.read_csv('Airline_customer_satisfaction.csv')
```

```python
review.head()
```

```python
review.describe()
```

```python
review.info()
```

```python
sns.displot(review['Flight Distance'])
```

```python
sns.displot(review['Age'])
```

```python
sns.displot(review['Arrival Delay in Minutes'])
```

```python
#Above plot have just small graph that is also around 0, so we are doing this now!
sns.displot(review['Arrival Delay in Minutes'])
plt.ylim(0,3000)
plt.xlabel('Arrival Delay in Minutes')
plt.ylabel('Count')
plt.title('Distribution of Arrival Delays (Filtered)')
plt.show()
```

```python
sns.displot(review['Cleanliness'])
```

```python
#Beautifying the display
sns.countplot(x='Cleanliness', data=review, palette='rainbow')
```

```python
sns.jointplot(review, x='Age', y='Cleanliness')
```

```python
sns.histplot(data=review, x='Cleanliness', hue='Class', multiple='dodge', bins=5)
plt.xlabel('Cleanliness Review')
plt.ylabel('Count of Reviews')
plt.title('Distribution of Cleanliness Reviews by Class')
plt.xticks([1, 2, 3, 4, 5])
plt.show()
```

```python
sns.histplot(data=review, x='On-board service', hue='Class', multiple='dodge', bins=5)
plt.xlabel('On-board Service')
plt.ylabel('Count of Reviews')
plt.title('Distribution of On-Board Service Reviews by Class')
plt.xticks([1, 2, 3, 4, 5])
plt.show()
```

```python
sns.boxplot(x='Class', y='On-board service', data=review)
plt.xlabel('Class of Travel')
plt.ylabel('Cleanliness Rating')
plt.title('Box Plot of On-board service Ratings by Class of Travel')
plt.show()
```

```python
sns.histplot(data=review, x='Cleanliness', hue='Customer Type', multiple='dodge', bins=5)
plt.xlabel('Cleanliness Review')
plt.ylabel('Count of Reviews')
plt.title('Distribution of Cleanliness Reviews by Customer Type')
plt.xticks([1, 2, 3, 4, 5])
plt.show()
```

```python
sns.boxplot(x='Class', y='Cleanliness', data=review)
plt.xlabel('Class of Travel')
plt.ylabel('Cleanliness Rating')
plt.title('Box Plot of Cleanliness Ratings by Class of Travel')
plt.show()
```

```python
sns.histplot(data=review, x='On-board service', hue='Customer Type', multiple='dodge', bins=5)
plt.xlabel('On-board Service')
plt.ylabel('Count of Reviews')
plt.title('Distribution of On-board Service Reviews by Customer Type')
plt.xticks([1, 2, 3, 4, 5])
plt.show()
```

```python
sns.countplot(x='satisfaction', data=review)
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
plt.title('Count of Satisfaction Levels')
plt.show()
```

```python
sns.countplot(x='Class', hue='satisfaction', data=review)
plt.xlabel('Class of Travel')
plt.ylabel('Count')
plt.title('Count of Satisfaction vs. Dissatisfaction Among Different Classes of Travel')
plt.show()
```
